Below code finds 5-star books from [this](https://books.toscrape.com/index.html) page and sorts them based on their prices in ascending order and displays
their names and prices on pandas DataFrame.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
fiveStarBooks = []
pageNo = 1
while(True:
    html_text = requests.get(f'http://books.toscrape.com/catalogue/page-{str(pageNo)}.html').text
    beautifulSoup = BeautifulSoup(html_text, 'lxml')
    if beautifulSoup.title.text == "404 Not Found": break
    books = beautifulSoup.find_all("article", class_ = "product_pod")
    for book in books:
        totalStars = book.find("p", class_ = "star-rating")["class"][1]
        if totalStars == "Five": fiveStarBooks.append(book)
    pageNo += 1
```

```python
booksNamesPrices = []
for book in fiveStarBooks:
    name = book.img["alt"]
    price = book.find("p", class_ = "price_color").text[1:]
    booksNamesPrices.append([name, price])
booksNamesPrices.sort(key = lambda book: book[1])
```

```python
pd.set_option('display.max_rows', len(booksNamesPrices))
df = pd.DataFrame(booksNamesPrices, columns = ["Name", "Price"])
df
```
